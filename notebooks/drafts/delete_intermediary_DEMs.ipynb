```python
import os
import glob
from tqdm.notebook import tqdm
```

```python
study_sites_path = '/Users/raineyaberle/Google Drive/My Drive/Research/CryoGARS-Glaciology/Advising/student-research/Alexandra-Friel/snow_cover_mapping_application/study-sites/'
site_names = sorted(os.listdir(study_sites_path))
site_names = [x for x in site_names if not x.startswith('.')]
site_names
```

```python
for site_name in tqdm(site_names):
    
    DEM_path = os.path.join(study_sites_path, site_name, 'DEMs')
    if os.path.exists(DEM_path):
        
        arcticdem_fn = os.path.join(DEM_path, site_name + '_ArcticDEM_clip.tif')
        arcticdem_geoid_fn = os.path.join(DEM_path, site_name + '_ArcticDEM_clip_geoid.tif')
        
        if os.path.exists(arcticdem_fn) and os.path.exists(arcticdem_geoid_fn):
            print(site_name)
            os.remove(arcticdem_fn)
            
```
